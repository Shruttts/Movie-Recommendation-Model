```python
import ast
```

ast.literal_eval() this function from ast module in python which stands for Abstract Syntax Trees will convert the list of string to the list

```python
def convert(obj):
  l=[]
  for i in ast.literal_eval(obj):
   l.append(i['name'])

  return l
```

```python
movie['genres']=movie['genres'].apply(convert)
```

```python
def convert5(obj):
  l=[]
  counter=0
  for i in ast.literal_eval(obj):
   l.append(i['name'])
   counter+=1
   if counter==5:
    break;
  return l
```

cv = CountVectorizer(...):
You are initializing a CountVectorizer object and assigning it to the variable cv.
max_features=5000: This parameter limits the number of features (i.e., the number of unique words) to the top 5000 most frequent words across all the documents in your dataset. This is a way to reduce the dimensionality of the data and focus on the most important words.
stop_words='English': This argument removes common English stop words (like "the", "is", "in", etc.) from the text during the vectorization process. It uses a predefined list of stop words available in Scikit-learn. These words are ignored when the text is transformed into vectors, as they don't contribute much to text analysis.

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

cv.fit_transform(new_df['tags']):

fit_transform() is a method in CountVectorizer that does two things:
fit(): It analyzes the text data in new_df['tags'] to build a vocabulary (a list of unique words) based on the text. The vocabulary will contain only the words that appear in the text, excluding stop words.
transform(): After fitting, it then transforms the text documents into a matrix of word counts. Each document is represented as a vector where each element corresponds to the frequency of a word from the vocabulary.
new_df['tags'] is assumed to be a column in your new_df dataframe containing text data, probably a list of tags or keywords for each document.
.toarray():

This converts the sparse matrix (default output of fit_transform()) into a dense NumPy array. A sparse matrix is memory-efficient and stores only the non-zero elements, but for easier manipulation or inspection, you may want to convert it into a dense array.

```python
cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

Will do stemming here
Stemming :
['loved', 'loving' , 'isncjcd']
it will give the main word love that is stemming

```python
import nltk
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
         y=[]
         for i in text.split():
           y.append(ps.stem(i))
         return " ".join(y)
```

```python
ps.stem('loved')
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
def recommend(movie):
  movie_index=new_df[new_df['title']==movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)) , reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```
